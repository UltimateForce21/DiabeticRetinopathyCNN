```python
import os
import shutil
from tqdm import tqdm

# Paths to the directories
train_sorted_path = '/content/gdrive/MyDrive/DR-dataset/diabetic-retinopathy-detection/train_sorted_resized'
aptos_train_images_path = '/content/gdrive/MyDrive/DR-dataset/aptos2019-blindness-detection/aptos_train_images_sorted_resized'
combined_path = '/content/gdrive/MyDrive/DR-dataset/aptos2019-kaggleDR-training-combined-resized'

# Create the combined directory if it doesn't exist
if not os.path.exists(combined_path):
    os.makedirs(combined_path)

# Classes in the dataset
classes = ['0', '1', '2', '3', '4']

# Combine images from both datasets
for cls in classes:
    # Create class directories in the combined folder if they don't exist
    class_combined_path = os.path.join(combined_path, cls)
    if not os.path.exists(class_combined_path):
        os.makedirs(class_combined_path)

    # Get images from train_sorted_resized
    train_class_path = os.path.join(train_sorted_path, cls)
    train_images = os.listdir(train_class_path)

    # Copy images from train_sorted_resized to combined folder
    for image in tqdm(train_images, desc=f'Copying {cls} from train_sorted_resized'):
        src = os.path.join(train_class_path, image)
        dst = os.path.join(class_combined_path, image)
        shutil.copy(src, dst)

    # Get images from aptos_train_images_sorted_resized
    aptos_class_path = os.path.join(aptos_train_images_path, cls)
    aptos_images = os.listdir(aptos_class_path)

    # Copy images from aptos_train_images_sorted_resized to combined folder
    for image in tqdm(aptos_images, desc=f'Copying {cls} from aptos_train_images_sorted_resized'):
        src = os.path.join(aptos_class_path, image)
        dst = os.path.join(class_combined_path, image)
        shutil.copy(src, dst)

print('Combining datasets completed.')
```

```python
import os
import cupy as cp
from imageio import imread
from cuml import SVC as cumlSVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, precision_score, recall_score, f1_score
from tqdm import tqdm

input_dir = '/content/aptos2019-kaggleDR-training-combined-resized'
categories = ['0', '1', '2', '3', '4']

data = []
labels = []

# Load images directly into CuPy arrays
for category_idx, category in enumerate(categories):
    category_path = os.path.join(input_dir, category)
    for file in os.listdir(category_path):
        img_path = os.path.join(category_path, file)
        img = imread(img_path)
        data.append(cp.asarray(img.flatten()))
        labels.append(category_idx)

images = np.array(data)
labels = np.array(labels)

# Use cuML's train_test_split
X_train, X_val, y_train, y_val = train_test_split(images, labels, test_size=0.15, shuffle=True, stratify=labels)

# Clear unused variables from system memory
del data
del labels
del images



# Define and train the SVM with RBF kernel
svm_gpu = cumlSVC(probability=True, kernel='rbf', C=1, gamma='scale')

# Batch size for incremental training
batch_size = 128
num_batches = int(cp.ceil(X_train.shape[0] / batch_size))

# Train the SVM model incrementally
for i in tqdm(range(num_batches), desc="Training Progress"):
    start_idx = i * batch_size
    end_idx = min(start_idx + batch_size, X_train.shape[0])
    X_batch = X_train[start_idx:end_idx]
    y_batch = y_train[start_idx:end_idx]
    svm_gpu.fit(X_batch, y_batch)

# Predict on the validation set
y_pred_gpu = svm_gpu.predict(X_val)

# Evaluate the performance
print("Validation Accuracy with GPU:", accuracy_score(cp.asnumpy(y_val), cp.asnumpy(y_pred_gpu)))
print(classification_report(cp.asnumpy(y_val), cp.asnumpy(y_pred_gpu)))
print("Precision:", precision_score(cp.asnumpy(y_val), cp.asnumpy(y_pred_gpu), average='weighted'))
print("Recall:", recall_score(cp.asnumpy(y_val), cp.asnumpy(y_pred_gpu), average='weighted'))
print("F1 Score:", f1_score(cp.asnumpy(y_val), cp.asnumpy(y_pred_gpu), average='weighted'))
```

```python
# Directory containing the resized images
input_dir = '/content/aptos2019-kaggleDR-training-combined-resized'
categories = ['0', '1', '2', '3', '4']

data = []
labels = []

for category_idx, category in enumerate(categories):
    category_path = os.path.join(input_dir, category)
    for file in tqdm(os.listdir(category_path), desc=f'Loading category {category}'):
        img_path = os.path.join(category_path, file)
        img = imread(img_path)
        img = resize(img, (128, 128))  # Resize to smaller dimension if necessary
        data.append(img.flatten())
        labels.append(category_idx)

# Convert the data and labels to numpy arrays
images = np.asarray(data)
labels = np.asarray(labels)
```

```python
# Perform PCA to reduce dimensionality
pca = PCA(n_components=100)  # Adjust the number of components as needed
images_pca = pca.fit_transform(images)

# Split the data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(images_pca, labels, test_size=0.15, shuffle=True, stratify=labels)
```

```python
# Get the best parameters and best estimator
best_params = grid_search.best_params_
best_svm = grid_search.best_estimator_

print("Best Parameters:", best_params)

# Train the SVM classifier with the best parameters on the training set
best_svm.fit(X_train, y_train)

# Predict on the validation set
y_pred = best_svm.predict(X_val)

print("\n\n")
# Evaluate the performance
print("Validation Accuracy:", accuracy_score(y_val, y_pred))
print("\n\n")
print(classification_report(y_val, y_pred))
print("\n\n")
print("Precision:", precision_score(y_val, y_pred, average='weighted'))
print("Recall:", recall_score(y_val, y_pred, average='weighted'))
print("F1 Score:", f1_score(y_val, y_pred, average='weighted'))
print("\n\n")
```

```python
# Define a custom transform to apply sharpening
class SharpenImage(object):
    def __call__(self, img):
        return F.adjust_sharpness(img, sharpness_factor=4)

# Define the transformations
augmentation_transforms = transforms.Compose([
    transforms.RandomApply([transforms.ColorJitter(brightness=0.3, contrast=0.2)], p=0.4),
    transforms.RandomApply([transforms.GaussianBlur(kernel_size=(5, 5))], p=0.3),
    SharpenImage()
])
```

```python
from PIL import Image, ImageOps
import os
import numpy as np
import random
from tqdm import tqdm  # Import the tqdm library

data = []
labels = []
augmented_data = []
augmented_labels = []

# Iterate over each category
for category_idx, category in enumerate(categories):
    category_path = os.path.join(input_dir, category)
    files = os.listdir(category_path)

    # Add a progress bar for the current category
    for file in tqdm(files, desc=f"Processing category {category}"):
        img_path = os.path.join(category_path, file)
        img = Image.open(img_path)
        img = img.resize((128, 128))  # Resize to smaller dimension if necessary
        img = augmentation_transforms(img)  # Apply augmentations

        # Add original image
        img_np = np.array(img).flatten()
        data.append(img_np)
        labels.append(category_idx)

        # Apply augmentations and add as new images for classes 1, 2, 3, 4
        if category_idx in [1, 2, 3, 4]:
            # Flip horizontally
            img_flip_horizontal = ImageOps.flip(img)
            img_flip_horizontal_np = np.array(img_flip_horizontal).flatten()
            augmented_data.append(img_flip_horizontal_np)
            augmented_labels.append(category_idx)

            # Flip vertically
            img_flip_vertical = ImageOps.mirror(img)
            img_flip_vertical_np = np.array(img_flip_vertical).flatten()
            augmented_data.append(img_flip_vertical_np)
            augmented_labels.append(category_idx)

            # Rotate randomly between 0 and 360 degrees
            img_rotate = img.rotate(random.uniform(0, 360))
            img_rotate_np = np.array(img_rotate).flatten()
            augmented_data.append(img_rotate_np)
            augmented_labels.append(category_idx)

# Combine original data and augmented data
data.extend(augmented_data)
labels.extend(augmented_labels)

# Convert to numpy arrays for further processing
data = np.array(data)
labels = np.array(labels)
```

```python
class DR_detection_neural_network (nn.Module):
  def __init__(self):
    super(DR_detection_neural_network,self).__init__()
    self.Conv1 = nn.Conv2d(3,8,3)
    self.Conv2 = nn.Conv2d(8,16,3)
    self.Conv3 = nn.Conv2d(16,32,3, stride=2)
    self.Conv4 = nn.Conv2d(32, 64, 3, stride=2)
    self.Conv5 = nn.Conv2d(64, 128, 3, stride=2)
    self.Conv6 = nn.Conv2d(128, 128, 3, stride=2)
    self.fc1 = nn.Linear(14*14*128, 256)
    self.fc2 = nn.Linear(256, 128)
    self.fc3 = nn.Linear(128, 5)
    self.Dropout = nn.Dropout(p=0.1)

  def forward(self,x):
    x = F.Relu(self.Conv1(x))
    x = F.Relu(self.Conv2(x))
    x = F.Relu(self.Conv3(x))
    x = F.Relu(self.Conv4(x))
    x = F.Relu(self.Conv5(x))
    x = F.Relu(self.Conv6(x))
    x = F.Relu(self.fc1(x))
    x = F.Relu(self.fc2(x))
    x = self.fc3(x)
    return x
```
